```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array([5])
y = np.array([10])

plt.scatter(x,y)
plt.show()
```

```python
x = np.array([20,10,15,20])
y = np.array([20,15,10,5])

plt.plot(x,y)
```

```python
plt.plot(x,y,marker="o")
plt.show()
```

```python
plt.plot(x,y,marker="*",mfc="r")
```

```python
plt.plot(x,y,marker="o",mfc="r",c="g",ls=":")
```

```python
plt.plot(x,y,marker="D",ms=20)
```

```python
plt.pie(x,labels=["a","b","c","d"])
plt.show()
```
